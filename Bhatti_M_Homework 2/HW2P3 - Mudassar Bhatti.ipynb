## Problem 3: Is the average of polls better than just one poll?

#### Problem 3(a)

Most undecided voters vote for one of the two candidates at the election. Therefore, the reported percentages underestimate the final value of both candidates. However, if we assume the undecided will split evenly, then the observed difference should be an unbiased estimate of the final difference. 

Add a new column to the `election` DataFrame containg the difference between Obama and Romeny called `Diff`. 

```python
election = pd.read_csv('http://elections.huffingtonpost.com/pollster/2012-general-election-romney-vs-obama.csv')
election.head()
```

```python
election['Undecided'].fillna(0, inplace=True)
election['Diff'] = abs(election['Obama'] - election['Romney'])
election['Diff'] = election['Diff'] / (election['Obama'] + election['Romney'] + election['Undecided'])
election.head()
```

#### Problem 3(c) 

Make a plot showing the differences by pollster where the pollsters are on the x-axis and the differences on the y-axis. 

```python
pollsters = list(set(election['Pollster'].values)) 
len(pollsters)
indexed_pollsters = {}
for i in range(len(pollsters)):
    indexed_pollsters[pollsters[i]] = i

indexed_pollsters
```

Is the *across poll* difference larger than the *between pollster* difference? 

```python
#your code here
X = [indexed_pollsters[pollster] for pollster in election['Pollster'].values]

fig = plt.figure(figsize=(16,4))
plt.scatter(X, election['Diff'].values, color ='r')
plt.xticks(X, election['Pollster'].values, rotation=90)
plt.ylabel("Percentage Difference")
plt.title("2012 Obama-Romney Poll")
plt.grid(True)
plt.show()
```

Answer: The number of polls show that the results between each one would be smaller then the differentiation bewteen all of the polls.
