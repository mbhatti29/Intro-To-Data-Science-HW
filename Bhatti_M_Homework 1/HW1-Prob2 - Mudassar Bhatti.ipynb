Transform the data set to have years as the rows and countries as the columns. Show the head of this data set when it is loaded. 

```python
#your code here
#income.index = income[income.columns[0]]
#income = income.drop(income.columns[0],axis =1)
#income = income.transpose()
#income.head()

income.index = income[income.columns[0]]
transform = income.drop(income.columns[0], axis =1)
transform.columns = map(lambda x: int(x), transform.columns)
transform = transform.T
transform.head(5)
```

#### Problem 2(b)

Graphically display the distribution of income per person across all countries in the world for any given year (e.g. 2000).  What kind of plot would be best? 

```python
years = 2000
plt.hist(np.log10(transform.ix[years].dropna().values), bins = 15)
plt.title('Year: %i' % years)
plt.xlabel('Income per person')
plt.ylabel('Scale')
plt.show()
```

#### Problem 2(c)

Write a function to merge the `countries` and `income` data sets for any given year. 

```python
"""
Function
--------
mergeByYear

Return a merged DataFrame containing the income, 
country name and region for a given year. 

Parameters
----------
year : int
    The year of interest

Returns
-------
a DataFrame
   A pandas DataFrame with three columns titled 
   'Country', 'Region', and 'Income'. 

Example
-------
>>> mergeByYear(2010)
"""
#your code here
def mergeByYear(yr):
    data1 = pd.DataFrame(transform.ix[yr].values, columns = ['Income'])
    data1['Country']=transform.columns
    total = pd.merge(data1, countries, on = ['Country'])
    total.Income = np.round(total.Income, 2)
    return total
mergeByYear(2010)
```

#### Problem 2(d) 

Use exploratory data analysis tools such as histograms and boxplots to explore the distribution of the income per person by region data set from 2(c) for a given year. Describe how these change through the recent years?

**Hint**: Use a `for` loop to consider multiple years.  

```python
#your code here
year0 = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
for yr in year0:
    df=mergeByYear(yr)
    df.boxplot('Income', by = 'Region', rot=90)
    plt.title("Year:" + str(yr))
    plt.ylabel("Income per person")
    plt.xlabel("Region")
    plt.ylim(10**2, 10.5 **5)
    plt.yscale('log')
```
